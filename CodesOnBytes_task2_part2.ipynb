# .....................................Simple Linear Regression...............................................

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X= data.iloc[:,:-1].values #Independent variable
y = data.iloc[:,-1].values #Dependent variable
```

```python
# Splitting the dataset in Training set and Test set
# so if the  whole dataset contains the 30 observations and test size=0.2 then 
#the split occurs such that 33.33% i.e 10 approx of all these observations will be in the test set.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state =1)
```

```python
print(X_train,y_train)
```

```python
print(X_test,y_test)
```

```python
# Training the simple linear regression model on the Training set
# Now from sklearn Library we will access Linear_model.
# From Linear_model import a class called LinearRegression to build simple Linear Regression model.
from sklearn.linear_model import LinearRegression
# Now we create an instance (object) of LinearRegression class. 
#Naming the object as 'regressor' and simply calling the class constructor LinearRegression(). 
# No parameters required.
regressor = LinearRegression()

# Now we have got a Linear Regression model but we have to train it on the training set.
# For this we have to connect it to the training set using fit() function using 'regressor' object.
# The fit() method will train this Simple Linear Regression model on the training set so,
# The fit() method parameters include: 
# 1. X_train: The matrix of features Independent variables in training set (x)
# 2. y train: The dependent variable in training set (y).
regressor.fit(X_train,y_train)
```

```python
y_pred1 = regressor.predict([[50]])
print(y_pred1)
y_pred2 = regressor.predict([[24]])
print(y_pred2)
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

```python
print('R squared:{:.2f}'.format(regressor.score(X,y)*100))
```

```python
regressor.score(X_test,y_test)
```
