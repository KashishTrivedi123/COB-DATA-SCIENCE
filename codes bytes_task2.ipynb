```python
def extract_season_count(duration):
    if "Season" in duration:
        return duration.split(" ")[0]
    else:
        return ""

def extract_duration_in_min(duration):
    if "Season" not in duration:
        return duration.split(" ")[0]
    else:
        return ""

# Apply the custom functions to create new columns
data['season_count'] = data['duration'].apply(extract_season_count)
data['duration_in_min'] = data['duration'].apply(extract_duration_in_min)
```
